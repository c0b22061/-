```python
import numpy as np
import matplotlib.pyplot as plt

from keras.models import Sequential,Model
from keras.layers import Dense, LeakyReLU, Input,BatchNormalization,Conv2D,Activation,MaxPooling2D,UpSampling2D,Reshape,Flatten
from keras.optimizers.legacy import Adam

from keras.datasets import mnist

n_learn = 1001
interval = 20
batch_size = 32
batch_half = 16
n_noise = 128
width = 28
height = 28
alpha = 0.2

sample_number = 8

(x_train, t_train), (x_test, t_test) = mnist.load_data()
print(x_train.shape)
print(x_test.shape)

x_train = x_train[t_train.flatten() == sample_number]
t_train = t_train[t_train.flatten() == sample_number]
print(x_train.shape)
print(t_train.shape)

plt.imshow(x_train[0],cmap="gray")
plt.title(t_train[0])
plt.show()

x_train = x_train/255*2-1

x_train = x_train.reshape(len(x_train),28,28,1)

generator = Sequential()
generator.add(Dense(1024,input_shape=(n_noise,),activation="tanh"))
generator.add(Dense(128*7*7,activation="tanh"))
generator.add(BatchNormalization())
generator.add(Reshape((7,7,128),input_shape=(7*7*128,)))
generator.add(UpSampling2D(size=(2,2)))
generator.add(Conv2D(64,(5,5),padding="same",activation="tanh"))
generator.add(UpSampling2D(size=(2,2)))
generator.add(Conv2D(1,(5,5),padding="same",activation="tanh"))

print(generator.summary())

discriminator = Sequential()
discriminator.add(Conv2D(64,(5,5),padding="same",input_shape=(28,28,1),activation="tanh"))
discriminator.add(MaxPooling2D(pool_size=(2,2)))
discriminator.add(Conv2D(128,(5,5),activation="tanh"))
discriminator.add(MaxPooling2D(pool_size=(2,2)))
discriminator.add(Flatten())
discriminator.add(Dense(1024,activation="tanh"))
discriminator.add(Dense(1,activation="sigmoid"))

discriminator.compile(loss="binary_crossentropy",optimizer=Adam(0.0002,0.5),metrics=["accuracy"])

print(discriminator.summary())

discriminator.trainable = False

noise = Input(shape=(n_noise,))
img = generator(noise)
reality = discriminator(img)

combined = Model(noise,reality)
combined.compile(loss="binary_crossentropy",optimizer=Adam(0.0002,0.5))
print(combined.summary())

for i in range(n_learn):
    gen_noise = np.random.normal(0,1,(batch_half,n_noise))
    gen_imgs = generator.predict(gen_noise)

    discriminator.train_on_batch(gen_imgs,np.zeros((batch_half,1)))

    img_id = np.random.randint(0,len(x_train),size=batch_half)
    real_imgs = x_train[img_id]

    discriminator.train_on_batch(real_imgs,np.ones((batch_half,1)))

    c_noise = np.random.normal(0,1,(batch_size,n_noise))
    combined.train_on_batch(c_noise,np.ones((batch_size,1)))

    print("n_learn:",i)
    if i % interval == 0:
      plt.figure(figsize=(2,2))
      output_img = gen_imgs[0].reshape(width,height)
      plt.imshow(output_img,cmap="gray")
      plt.show()
plt.figure(figsize=(10,10))
for i in range(0,8):
  plt.subplot(1,8,i+1)
  output_img = gen_imgs[i].reshape(width,height)
  output_img = (output_img+1)/2
  plt.imshow(output_img,cmap="gray")
  plt.tick_params(labelbottom=False,labelleft=False,bottom=False,left=False)
plt.show()
```
